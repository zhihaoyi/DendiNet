```python
# return to main directory
import os
import sys

current_directory = os.getcwd()
main_dir = os.path.dirname(current_directory)
sys.path.append(main_dir)
```

```python
import torch
import matplotlib.pyplot as plt
import pickle

from utils.F_train_validate import validate
from utils.loss import rmse

# cuda
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'avaiable device: {device}')
```

```python
# load the model
model = torch.load(main_dir + '/model_save/gru_params.pth')
```

### US06

```python
# load the dataloader
with open(main_dir + '/dataloader/US06_test_loader.pkl', 'rb') as file:
    US06_test_loader = pickle.load(file)
```

```python
# validate the test dataset
US06_test_loss_epoch, US06_test_target_log, US06_test_pred_log = validate(model, US06_test_loader, rmse, device=device)
print(f'US06 testing loss: {US06_test_loss_epoch}')
```

```python
plt.plot(US06_test_pred_log, label='prediction')
plt.plot(US06_test_target_log, label='target')
plt.title('GRU Prediction - US06')
plt.ylabel('Ah')
plt.xlabel('frames')
plt.legend()
plt.show()
```
